```python
traces = list(traces)
```

```python
import pyevtk
from evtk import hl
def trace_to_vtk_frame(filename,trace): 
    """export all info within a trace to a vtu for viewing in paraview
    a trace is a reconstructed frame with the 3d positions of all moving
    particles in a numpy array associated with a particular time step"""
    x,y,z = trace.T
    x = np.ascontiguousarray(x)
    y = np.ascontiguousarray(y)
    z = np.ascontiguousarray(z)
    dats = {'diameter': np.ones_like(x)*0.5} # particles are 0.5cm diameter
    hl.pointsToVTK(filename,x,y,z,dats)
    
def tracks_to_vtk_frames(traces,path):
    for i,trace in enumerate(traces):
        filename = path+'trace-%06d'%i
        trace_to_vtk_frame(filename,trace)
        print('frame %d saved'%i)
```

```python
path = '../trajectories/vtkvis/'
```
